将所需要的数据保存为单独的csv文件
1. 全体数据的序号、身高（cm）、体重（kg）、BMI
2. 抽取1000个身高大于173cm的人的序号、身高（cm）、体重（kg）、BMI
3. 抽取1000个身高小于等于173cm的人的序号、身高（cm）、体重（kg）、BMI

```python
RANDOM_STATE = 2023
```

```python
converted_csv = pd.read_csv('../data/SOCR-HeightWeight_converted.csv')
```

```python
converted_csv.head(5)
```

```python
# 1. 全体数据的序号、身高（cm）、体重（kg）、BMI
converted_csv[['Index', 'Height(cm)', 'Weight(kg)', 'BMI']].to_csv('../data/DataAll.csv', index=False)
```

```python
# 2. 抽取5000个身高大于173cm的人的序号、身高（cm）、体重（kg）、BMI
num_samples = 5000
sampled_taller = converted_csv[converted_csv['Height(cm)'] > 173].sample(num_samples, replace=False, random_state=RANDOM_STATE)
sampled_taller.head()
```

```python
sampled_taller[['Index', 'Height(cm)', 'Weight(kg)', 'BMI']].to_csv('../data/DataTaller.csv', index=False)
```

```python
# 3. 抽取1000个身高小于等于173cm的人的序号、身高（cm）、体重（kg）、BMI
sampled_shorter = converted_csv[converted_csv['Height(cm)'] <= 173].sample(num_samples, replace=False, random_state=RANDOM_STATE)
sampled_shorter.head()
```

```python
sampled_shorter[['Index', 'Height(cm)', 'Weight(kg)', 'BMI']].to_csv('../data/DataShorter.csv', index=False)
```
